```python
import matplotlib.pyplot as plt

with open('D:\Spoter_transformer_SLR\selfPosValues_RandomlyIni.txt', 'r') as file:
    data = [list(map(float, line.split(','))) for line in file]

transposed_data = list(map(list, zip(*data)))

for i, row in enumerate(transposed_data):
    plt.plot(row)

plt.xlabel('Epoch')
plt.ylabel('Positional encoding value')
plt.title('Change in training - randomly initialised PE')

# Add legend
# plt.legend()

plt.show()
```
